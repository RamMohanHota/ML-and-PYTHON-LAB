```python
#Create a simple line plot of y = 2x + 3 for x in the range 010. Add appropriate title, x-label, y-label, and display the grid.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100) # Generate 100 points between 0 and 10
y = 2 * x + 3

plt.plot(x, y)
plt.title('Line Plot of y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#Plot y = x² for x between-10 and 10 using: A red dashed line (r--), Line width of 2, Marker at each data point.
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100) # Generate 100 points between -10 and 10
y = x**2

plt.plot(x, y, 'r--', linewidth=2, marker='o') # 'r--' for red dashed line, linewidth=2, marker='o' for markers
plt.title('Plot of y = x²')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
#Create a 1Ö2 subplot:Left subplot Plot sin(x)Right subplotPlot cos(x)for x between 0 and 2 .
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 100) # Generate 100 points between 0 and 2pi
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(1, 2, 1) # 1 row, 2 columns, 1st plot
plt.plot(x, y_sin)
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)

plt.subplot(1, 2, 2) # 1 row, 2 columns, 2nd plot
plt.plot(x, y_cos)
plt.title('cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)

plt.tight_layout() # Adjust layout to prevent titles/labels overlapping
plt.show()
```

```python
#Generate 50 random x and y values and plot a scatter plot. Use a di erent color and marker size for better visualization.
import matplotlib.pyplot as plt
import numpy as np

# Generate 50 random x and y values
np.random.seed(42) # for reproducibility
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot with different color and marker size
plt.scatter(x, y, color='purple', s=100, alpha=0.7) # s is marker size, alpha is transparency

plt.title('Random Scatter Plot')
plt.xlabel('Random X Values')
plt.ylabel('Random Y Values')
plt.grid(True)
plt.show()
```

```python
#6 Given sales data for products: Products = [A,B,C,D], Sales = [120, 150, 90,200] Create a bar plot of sales per product with proper labels.
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Sales data from Question 6
products = ['A', 'B', 'C', 'D']
sales = [120, 150, 90, 200]

# Create a pandas DataFrame to easily sort the data
sales_df = pd.DataFrame({'Product': products, 'Sales': sales})

# Sort the DataFrame by sales in descending order
sales_df_sorted = sales_df.sort_values(by='Sales', ascending=False)

# Create the horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(sales_df_sorted['Product'], sales_df_sorted['Sales'], color='lightgreen')

# Add title and labels
plt.title('Sales per Product (Descending Order)')
plt.xlabel('Sales')
plt.ylabel('Product')

# Add grid
plt.grid(axis='x', linestyle='--')

# Show the plot
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
data = np.random.normal(mean, std_dev, 1000)

# Create the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')

# Add title and labels
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show gridlines
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
#labels = [Python, Java, C++, JavaScript] usage = [40, 30, 15, 15] Create a pie chart with: Explode the largest slice, Show percentages on each slice, Add a shadow.
import matplotlib.pyplot as plt
import numpy as np

# Data for the pie chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
usage = [40, 30, 15, 15]

# Find the index of the largest slice to explode it
explode = [0] * len(usage)
largest_slice_index = usage.index(max(usage))
explode[largest_slice_index] = 0.1  # Explode the largest slice

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(usage, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)

# Add a title
plt.title('Programming Language Usage')

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the plot
plt.show()
```
