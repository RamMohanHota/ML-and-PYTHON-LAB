```python
def random_centroids(data,k):
  n_samples,n_features=data.shape
  random_sample_idxs=np.random.choice(n_samples,k,replace=False)
  centroids=[]
  for i in random_sample_idxs:
    centroid=data.iloc[i]
    centroids.append(centroid)
  centroids=pd.concat(centroids,axis=1)
  cluster_no=[id for id in range(k)]
  centroids.columns=cluster_no
  return centroids
```

```python
def get_labels(data, centroids):
    distances = centroids.apply(lambda x: np.sqrt(((data - x) ** 2).sum(axis=1)))
    labels = distances.idxmin(axis=1)
    return labels
```

```python
def random_centroids(data,k):
  numeric_data = data.select_dtypes(include=np.number) # Select only numeric columns
  n_samples,n_features=numeric_data.shape
  if n_samples < k:
      raise ValueError("Number of samples in data is less than k")
  random_sample_idxs=np.random.choice(n_samples,k,replace=False)
  centroids = numeric_data.iloc[random_sample_idxs].T # Sample rows and transpose
  cluster_no=[id for id in range(k)]
  centroids.columns=cluster_no
  return centroids
```

```python
def new_centroids(data,labels,k):
  centroids=data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
  return centroids
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

```python
def plot_clusters(data,labels,centroids,iteration):
  pca=PCA(n_components=2)
  data_2d=pca.fit_transform(data)
  centroids_2d=pca.transform(centroids.T)
  clear_output(wait=True)
  plt.title(f'Iteration {iteration}')
  plt.scatter(x=data_2d[:,0],y=data_2d[:,1],c=labels)
  plt.scatter(x=centroids_2d[:,0],y=centroids_2d[:,1])
  plt.show()
```
